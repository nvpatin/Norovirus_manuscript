## This script takes a raw output file from vsearch and formats it for the next step (merging outputs from multiple files for downstream analyses).

## It only does one file at at a time. Could be edited to run in a loop on all files in a directory if they are formatted the same way.

## For the Norovirus manuscript this was run before "vsearch_out_merge"

```python
import pandas as pd
```

```python
cd Raw
```

```python
path1="4_1_vsearch.b6"
```

```python
df = pd.read_csv(path1, header=None, sep="\t")
```

```python
df2 = pd.concat([df[0].str.split(';', expand=True), df[[1]]], axis=1)
df2.columns = ('ID','Count','Taxonomy')
df2
```

### Remove all characters before the phylum name

```python
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Bacteroidetes', 'Bacteroidetes')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Firmicutes', 'Firmicutes')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Actinobacteria', 'Actinobacteria')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Proteobacteria', 'Proteobacteria')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Patescibacteria', 'Patescibacteria')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Tenericutes', 'Tenericutes')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Rokubacteria', 'Rokubacteria')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Chloroflexi', 'Chloroflexi')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Cyanobacteria', 'Cyanobacteria')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Synergistetes', 'Synergistetes')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Fibrobacteres', 'Fibrobacteres')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Verrucomicrobia', 'Verrucomicrobia')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Chlamydiae', 'Chlamydiae')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Planctomycetes', 'Planctomycetes')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Kiritimatiellaeota', 'Kiritimatiellaeota')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*GN01', 'GN01')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Gemmatimonadetes', 'Gemmatimonadetes')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Spirochaetes', 'Spirochaetes')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Atribacteria', 'Atribacteria')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Firestonebacteria', 'Firestonebacteria')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Deinococcus', 'Deinococcus')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Elusimicrobia', 'Elusimicrobia')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Epsilonbacteraeota', 'Epsilonbacteraeota')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Acidobacteria', 'Acidobacteria')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Cloacimonetes', 'Cloacimonetes')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Fusobacteriia', 'Fusobacteriia')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Lentisphaerae', 'Lentisphaerae')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Omnitrophicaeota', 'Omnitrophicaeota')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Margulisbacteria', 'Margulisbacteria')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Halanaerobiaeota', 'Halanaerobiaeota')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Aegiribacteria', 'Aegiribacteria')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Zixibacteria', 'Zixibacteria')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Nitrospirae', 'Nitrospirae')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*FBP', 'FBP')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Marinimicrobia', 'Marinimicrobia')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Dependentiae', 'Dependentiae')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Aquificae', 'Aquificae')
df2['Taxonomy'] = df2['Taxonomy'].str.replace(r'.*Thermotogae', 'Thermotogae')
df2
```

```python
type(df2)
```

```python
df2['Count'] = df2['Count'].astype('str')
df2['Count'] = df2['Count'].str.replace('size=', '', regex=False)
df2['Count'] = df2['Count'].astype('int')
df2 = df2.drop(['ID'], axis=1)
```

### Combine all rows with the same taxonomy string and sum their counts

```python
df3 = df2.groupby(['Taxonomy']).sum()
df3 = df3.rename(columns={"Count": "4_1"})
df3
```

```python
df3.to_csv("4_1_pdformat.csv")
```
